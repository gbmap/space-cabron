```python
import tensorflow as tf
import pandas as pd
import numpy as np

from model.vae import VAE
```

```python
file = ".\\Datasets\\synths_metadata.csv"
data = pd.read_csv(file)
data.sample(1)

model = VAE(6, data.shape[1], 2)
load_status = model.load_weights(".\\Models\\basses_weights")
```

```python
z = model.reparameterize(tf.constant([[0.5, 0.5, 0.5]]), tf.constant([[0.5,0.5,0.5]]))
generated_synth = tf.round(model.decode(z))
pd.DataFrame(generated_synth, columns=data.columns)
```

```python
%%capture
import random 

synths_to_generate = 50
new_table = pd.DataFrame(columns=data.columns)
for i in range(synths_to_generate):
    mean = tf.random.normal(shape=(1,3), mean=random.random()*2.0-1.0, stddev=2.0)
    logvar = tf.random.normal(shape=(1,3), mean=random.random()*2.0-1.0, stddev=2.0) 
    z = model.reparameterize(mean, logvar)
    new_synth_table = pd.DataFrame(tf.round(model.decode(z)), columns=data.columns)
    new_table = new_table.append(new_synth_table)
```
